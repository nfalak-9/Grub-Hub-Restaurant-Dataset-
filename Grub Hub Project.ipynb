```python
# Achieve a 20% reduction in memory usage by dropping irrelevant columns and optimizing data types
irrelevant_columns = [ 'Longitude', 'Latitude','Locality Verbose']
df.drop(columns=irrelevant_columns, inplace=True)
```

```python
df
```

```python
# Convert 'Price range' to categorical to optimize memory usage
df['Price range'] = df['Price range'].astype('category')
df
```

# Univariant Analysis


```python
# Analysing costumer has table bookings or not
sns.countplot(x='Has Table booking', data=df)


# Obervation : Most of the customers has no booking  
```

```python
# Analysing costumer has online delivery or not
sns.countplot(x='Has Online delivery', data=df)

# Observation : Some of the Costumer has Online Delivery
```

```python
# Analysing the no of delivery in present
sns.countplot(x='Is delivering now', data=df)

# Observation : Very Less Deliveries
```

```python
# Percentage count of Price range
df['Price range'].value_counts().plot(kind='pie', autopct = '%.2f')

#Observation : Most of the costumers belongs to Price Range 1
```

```python
# Percentage Count of Ratings

df['Aggregate rating'].value_counts().plot(kind='pie', autopct = '%.2f')

#Observation : 22.51% costumers given 0 ratings
```

```python
# Pie chart displaying the top 5 cuisines
top_cuisines = df['Cuisines'].value_counts().head(5)
plt.figure(figsize=(8, 6))
top_cuisines.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 Cuisines')
plt.ylabel('')
plt.show()
```
